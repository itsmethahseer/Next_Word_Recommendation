## Next_Word_Prediction

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.layers import LSTM,Dense,Embedding
from tensorflow.keras.models import Sequential
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.optimizers import Adam
import pickle
import numpy as np
import os
```

```python
file = open("book.txt","r",encoding="utf8")
```

```python
lines = []
for i in file:
    lines.append(i)
```

```python
lines
```

```python
### i'm going to convert the list into string
data = ""
for i in lines:
    data = ' '.join(lines)
```

```python
data
```

```python
#### Replace unnecessory characters with space
data = data.replace('\n','').replace('\r','').replace('\ufeff','').replace('"','')
```

```python
data
```

```python
#### Remove the unnecessory spaces by converting into a array and again into String
data = data.split()
```

```python
data
```

```python
data = ' '.join(data)
```

```python
data[:500]
```

```python
len(data)
```

```python
tokenizer = Tokenizer()
tokenizer.fit_on_texts([data])
```

```python
####for prediction function i'm going to save my tokenizer model
pickle.dump(tokenizer,open("my_tokenizer","wb"))
```

```python
sequence_of_data = tokenizer.texts_to_sequences([data])[0]
```

```python
sequence_of_data[:20]
```

Now each word will get the unique number

```python
len(sequence_of_data)
```

```python
vocab_size = len(tokenizer.word_index)+1
vocab_size
```

#### i'm going to train my model by passing three words are input and predict next word. i select 4 words as one set in which 3 of them are input and 4 th one is output

```python
sequences = []
for i in range(3,len(sequence_of_data)):
    words = sequence_of_data[i-3:i+1]
    print(words)
    break
    
```

```python
sequences = []
for i in range(3,len(sequence_of_data)):
    words = sequence_of_data[i-3:i+1]
    sequences.append(words)
sequences[:10]
```

```python
print('length of the sequences are',len(sequences))
```

Now i'm going to convert my sequences array into X and Y for training my model

```python
X = []
Y = []
for i in sequences:
    X.append(i[0:3])
    Y.append(i[3])
X = np.array(X)
Y = np.array(Y)
print("inputs :",X[:10])
print("responses :",Y[:10])
```

```python
### Now i'm going to convert my target variable into onehotencoded vectors
```

```python
y = to_categorical(Y,num_classes=vocab_size)
y
```

The to_categorical function is commonly used to convert integer-encoded categorical variables into one-hot encoded vectors. It is often used in machine learning tasks, such as multi-class classification, where the target variable needs to be represented as a binary matrix.

### creating the model

```python
model = Sequential()
model.add(Embedding(vocab_size, 10, input_length=3))
model.add(LSTM(1000, return_sequences=True))
model.add(LSTM(1000))
model.add(Dense(1000, activation="relu"))
model.add(Dense(vocab_size, activation="softmax"))
model.summary()
```

#### train the model

```python
from tensorflow.keras.callbacks import ModelCheckpoint
 
checkpoint = ModelCheckpoint("next_words.h5", monitor='loss', verbose=1, save_best_only=True)
model.compile(loss="categorical_crossentropy", optimizer=Adam(learning_rate=0.001))
model.fit(X, y, epochs=10, batch_size=64, callbacks=[checkpoint])
```

batch_size = 64 means that model will take first 64 samples to train the model, then takes next 64 and go on
